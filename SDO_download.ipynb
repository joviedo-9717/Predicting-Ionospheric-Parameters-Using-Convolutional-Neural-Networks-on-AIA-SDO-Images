```python
# Please dont run this code again as the data cleaning was done
# separately and remotely.

"""
import os
import requests
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
from tqdm import tqdm

from google.colab import drive
drive.mount('/content/drive')

# Parameters
start_date_str = '2024-01-01'
end_date_str = '2024-05-30'
target_time_strs = ['06:00:00', '12:00:00', '18:00:00']
wavelengths = ['0094', '0131', '0171', '0193', '0211', '0304']
resolution = '512'

base_dir = '/content/drive/MyDrive/CNN_SDO/SDO_data'
os.makedirs(base_dir, exist_ok=True)
for wl in wavelengths:
    os.makedirs(os.path.join(base_dir, wl), exist_ok=True)

start_date = datetime.strptime(start_date_str, '%Y-%m-%d')
end_date = datetime.strptime(end_date_str, '%Y-%m-%d')
target_times = [datetime.strptime(t, '%H:%M:%S').time() for t in target_time_strs]

dates = []
current_date = start_date
while current_date <= end_date:
    dates.append(current_date)
    current_date += timedelta(days=1)

for current_date in tqdm(dates, desc="Downloading SDO images"):
    date_path = current_date.strftime('%Y/%m/%d')
    date_str = current_date.strftime('%Y%m%d')
    url_base = f'https://sdo.gsfc.nasa.gov/assets/img/browse/{date_path}/'

    response = requests.get(url_base)
    if response.status_code != 200:
        continue

    soup = BeautifulSoup(response.text, 'html.parser')
    files = [a['href'] for a in soup.find_all('a') if a['href'].endswith('.jpg')]

    for wl in wavelengths:
        pattern = f'_{resolution}_{wl}.jpg'
        wl_files = [f for f in files if pattern in f and f.startswith(date_str)]

        if not wl_files:
            continue

        # For each target time, find and save the closest image
        for target_time in target_times:
            closest_file = None
            min_time_diff = timedelta.max
            for f in wl_files:
                time_str = f.split('_')[1]
                img_time = datetime.strptime(time_str, '%H%M%S').time()
                diff = abs(datetime.combine(datetime.min, img_time) - datetime.combine(datetime.min, target_time))
                if diff < min_time_diff:
                    min_time_diff = diff
                    closest_file = f

            # Save with time in the filename to avoid overwriting (optional)
            save_path = os.path.join(base_dir, wl, f"{date_str}_{target_time.strftime('%H%M%S')}_{resolution}_{wl}.jpg")
            os.makedirs(os.path.dirname(save_path), exist_ok=True)
            if not os.path.exists(save_path):
                img_url = url_base + closest_file
                img_data = requests.get(img_url).content
                with open(save_path, 'wb') as f:
                    f.write(img_data)

print("All downloads completed.")
"""
```
